```python
jsonFileObject = {
            'sourceDashboards' : [],
            'sourceCollections' : [],
            'sourceCards' : [],
            'sourceDashboardCards' : []
        }
url = 'http://13.233.43.107:3000/api'
```

```python
sourceDashboard = requests.get(f'{url}/dashboard/', headers=headers).json()
jsonFileObject['sourceDashboards']  = sourceDashboard
sourceCollections = requests.get(f'{url}/collection/', headers=headers).json()
jsonFileObject['sourceCollections'] = sourceCollections
sourceCards = requests.get(f'{url}/cards', headers=headers).json()
jsonFileObject['sourceCards'] = sourceCards
for i in sourceDashboard:
    sourceDashboardCards = requests.get(f"{url}/dashboard/{i['id']}", headers=headers).json()
    jsonFileObject['sourceDashboardCards'].append(sourceDashboardCards)

jsonFileObject['sourceDashboardCards']
```
